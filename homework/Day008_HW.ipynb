### 1. Dcard 網址： https://www.dcard.tw/f

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.dcard.tw/f'

r = requests.get(url)

r.encoding = 'utf-8'
print(r.text[0:3000])
```

```python
print('Request 取回之後該怎麼取出資料，資料型態是什麼？ =>')
print(type(r))
```

```python
print('為什麼要使用 BeatifulSoup 處理？處理後的型態是什麼？ => ')

soup =BeautifulSoup(r.text,"html5lib")
print(type(soup))
soup
```

### 2. 知乎： https://www.zhihu.com/explore

```python
url = 'https://www.zhihu.com/explore'
r = requests.get(url)
r.encoding = 'utf-8'

print(r.text[0:600])
```

### 3. 觀察一下知乎回來的資料好像有點怪怪的，該怎麼解決？

```python
import requests
url = 'https://www.zhihu.com/explore'

headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.92 Safari/537.36'}
r = requests.get(url,headers=headers)

r.encoding = 'utf-8'


print(r.text[0:600])

```
