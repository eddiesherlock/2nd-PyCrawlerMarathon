## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？

import requests, json
url = 'https://www.dcard.tw/_api/forums/pet/posts?popular=true'
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_2) AppleWebKit/605.1.15 (KHTML, like Gecko) \
                         Version/13.0.4 Safari/605.1.15"}
res = requests.get(url, headers=headers)
jdata = json.loads(res.text)
print(len(jdata))
json.loads(res.text)
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」

comment_sum = 0
like_sum = 0
for t in jdata:
    title = t['title']
    t_time = t['createdAt']
    t_commentCount = t['commentCount']
    t_like = t['likeCount']
    print('==========================')
    print('標題:', title)
    print('貼文時間:', t_time)
    print('留言人數:', t_commentCount)
    print('按讚人數:', t_like)
    comment_sum += t_commentCount
    like_sum += t_like
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」
print('==========================')
print('平均留言人數:', comment_sum / len(jdata))
print('平均按讚人數:', like_sum / len(jdata))
```
