2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€

```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ

import requests, json
url = 'https://www.dcard.tw/_api/forums/pet/posts?popular=true'
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_2) AppleWebKit/605.1.15 (KHTML, like Gecko) \
                         Version/13.0.4 Safari/605.1.15"}
res = requests.get(url, headers=headers)
jdata = json.loads(res.text)
print(len(jdata))
json.loads(res.text)
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€

comment_sum = 0
like_sum = 0
for t in jdata:
    title = t['title']
    t_time = t['createdAt']
    t_commentCount = t['commentCount']
    t_like = t['likeCount']
    print('==========================')
    print('æ¨™é¡Œ:', title)
    print('è²¼æ–‡æ™‚é–“:', t_time)
    print('ç•™è¨€äººæ•¸:', t_commentCount)
    print('æŒ‰è®šäººæ•¸:', t_like)
    comment_sum += t_commentCount
    like_sum += t_like
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€
print('==========================')
print('å¹³å‡ç•™è¨€äººæ•¸:', comment_sum / len(jdata))
print('å¹³å‡æŒ‰è®šäººæ•¸:', like_sum / len(jdata))
```
